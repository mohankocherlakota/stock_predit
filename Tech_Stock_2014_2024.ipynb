```python
import pandas as pd
import os

# List of file paths (replace with your actual paths)
file_paths = [
    'AAPL_2014_2024.csv',
    'ACN_2014_2024.csv',
    'ADBE_2014_2024.csv',
    'AMD_2014_2024.csv',
    'AVGO_2014_2024.csv',
    'CRM_2014_2024.csv',
    'CSCO_2014_2024.csv',
    'NVDA_2014_2024.csv',
    'ORCL_2014_2024.csv'
]

# Combine all CSVs into one DataFrame with an additional column
dataframes = []
for file in file_paths:
    # Extract the identifier from the file name
    identifier = os.path.basename(file).split('_')[0]
    # Read CSV and add a new column
    df = pd.read_csv(file)
    df['Source'] = identifier
    dataframes.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dataframes)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('combined_stocks_2014_2024.csv', index=False)

print("CSV files combined successfully into 'combined_2014_2024.csv' with a new 'Source' column")
```
